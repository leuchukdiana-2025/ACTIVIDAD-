```python
import pandas as pd
from sklearn.preprocessing import StandardScaler
from sklearn.neighbors import KNeighborsClassifier

df = pd.read_csv('winequality-red.csv')
df.info()
df.head()
```

```python
x = df[['alcohol', 'volatile acidity', 'chlorides']]
print (x)

y = df['quality']
scaler = StandardScaler() #Escalo el modelo
XScaled = scaler.fit_transform(x)

Kn=KNeighborsClassifier(n_neighbors=5) #Creacion de modelo
Kn.fit(XScaled, y )
```

```python
input = [[ 11.0, 0.300, 0.060]] #DOY VALORES PARA PREDECIR
input_df = pd.DataFrame(input, columns=['alcohol', 'volatile acidity', 'chlorides'])
inputScaler = scaler.transform(input_df)
output= Kn.predict(inputScaler) #PREDICCION HAY QUE ASOCIARLA A UNA SALIDA

print("La calidad del vino es: " , output[0])
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# ================================
# Gráfico de dispersión Alcohol vs Quality
# ================================
plt.figure(figsize=(6,4))
sns.scatterplot(data=df, x="alcohol", y="volatile acidity", hue="quality", palette="viridis")
plt.title("Calidad del vino")
plt.xlabel("Porcentaje de alcohol")
plt.ylabel("Acidez volátil")
plt.show()
```
